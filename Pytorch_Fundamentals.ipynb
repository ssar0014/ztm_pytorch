```python
import torch

print(torch.__version__)
```

# Torch Tensors

```python
# scalar - 1 dimenstional vector/tensor
type(torch.tensor(120))
```

```python
# 2 dimensional vector/tensor
torch.tensor([[120, 12],
              [144, 12]])
```

```python
# 3 dimensional vector/tensor
matrix = torch.tensor([
    [   [1,2,3],
        [3,4,5],
        [5,6,7]
    ],
    [   [1,2,3],
        [3,4,5],
        [5,6,7]
    ],
    [   [1,2,3],
        [3,4,5],
        [5,6,7]
    ]
])
```

```python
print(f"Matrix Dimensions: {matrix.ndim}\nMatrix Shape: {matrix.shape}")
```

# Random Tensors

```python
torch.rand(3,3,3)
```

```python
# create a random image generated by random tensors
image = torch.rand(255,255,3)
```

```python
from matplotlib import pyplot as plt
plt.imshow(image)
```

# One hot tensors (and other types)

```python
# All zeros tensor
torch.zeros(3,3,3)
```

```python
# all ones
torch.ones(3,3,3)
```

```python
# diagonal matrix - non square
torch.eye(4,3)
```

```python
# diagonal matrix - square
torch.eye(3)
```

```python
b = image * torch.eye(255, 3)
```
